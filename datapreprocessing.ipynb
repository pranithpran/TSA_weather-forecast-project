```python
import pandas as pd
import numpy as np
from datetime import timedelta

def data_generator(xt, new_df, k):
    # Check necessary columns
    if "Formatted Date" not in new_df.columns:
        raise KeyError("'Formatted Date' column is missing in new_df.")
    if "Date" not in new_df.columns:
        raise KeyError("'Date' column is missing in new_df.")
    
    # Ensure 'Formatted Date' is datetime
    new_df["Formatted Date"] = pd.to_datetime(new_df["Formatted Date"], errors="coerce")
    
    # Ensure 'Hour' column exists
    if "Hour" not in new_df.columns:
        new_df["Hour"] = new_df["Formatted Date"].dt.hour
    
    # Filter rows for the given date
    g = new_df[new_df["Date"] == xt].copy()
    if g.empty:
        raise ValueError(f"No data found for the specified date: {xt}")
    
    if "Hour" not in g.columns:
        g["Hour"] = g["Formatted Date"].dt.hour
    
    # Generate missing hours
    q = pd.Series(range(0, 24))
    p = q[~(q.isin(g["Hour"].unique()))]
    md = {}
    pw = []

    for i in range(len(p)):
        gd = {}
        for j in g.columns[1:len(g.columns) - 2]:
            d_5 = [xt - timedelta(days=n) for n in range(1, k)]
            gen = new_df[(new_df["Date"].isin(d_5)) & (new_df["Hour"] == p.iloc[i])]

            if not gen.empty:
                if pd.api.types.is_object_dtype(g[j]):
                    gd[j] = gen[j].mode()[0] if not gen[j].mode().empty else None
                else:
                    gd[j] = gen[j].median()
            else:
                gd[j] = None

        md[str(xt) + " " + str(p.iloc[i])] = gd
        pw.append(p.iloc[i])

    fr = pd.DataFrame(md).T.reset_index(drop=True)
    col = fr.columns
    d1 = np.array(fr)
    mr = pd.DataFrame(d1, columns=col)
    mr["Hour"] = pw
    mr["Date"] = xt
    return mr
```
