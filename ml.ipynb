```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Weather Cleaned Data.csv').head(5000)
df
```

```python
df.isnull().sum()
```

```python

df['Precip Type']=df['Precip Type'].bfill()
```

```python
df.isnull().sum()
```

```python
df['Temperature (C)'].corr(df['Apparent Temperature (C)'])
```

```python
df[['Temperature (C)','Humidity','Wind Speed (km/h)','Visibility (km)','Pressure (millibars)','Wind Bearing (degrees)']].corr()
```

```python
df.drop("Apparent Temperature (C)",axis=1,inplace=True)
```

```python
df = df.sort_values(by=['Date', 'Hour'], ascending=[True, False])
```

```python
df
```

```python
plt.plot(df['Humidity'])
```

```python
from statsmodels.graphics.tsaplots import plot_acf
```

```python
plot_acf(df["Humidity"],lags=30)
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
result = adfuller(df['Humidity'])
```

```python
result
```

```python
sns.lineplot(data=df, x=df.index, y='Humidity')

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Humidity')
plt.title('Humidity Line Plot')

plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
s = SARIMAX(df['Humidity'],seasonal_order=(1,1,1,13))
```

```python
s = s.fit()
s
```

```python
import pickle
with open('model.pickle','wb') as model_file:
    pickle.dump(s,model_file)
```
